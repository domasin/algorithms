```polyglot-notebook
#load "Asympt.fsx"
open Asympt
```

```polyglot-notebook
// quadratic

let x, y =
    [1..10..100]
    |> List.map (fun n ->
        let stopWatch = System.Diagnostics.Stopwatch.StartNew()
        let quad = quadraticTime n
        stopWatch.Stop()
        (n, stopWatch.Elapsed.TotalMilliseconds))
    |> List.unzip

Chart.Line(x, y)
|> Chart.withXAxisStyle("numInput")
|> Chart.withYAxisStyle("milliseconds")
|> Chart.withTitle("n^2 time")
|> Chart.withLineStyle(Width=2.0, Dash=StyleParam.DrawingStyle.Solid)
```

```polyglot-notebook
// logaritmic

let x, y =
    [1..10..10000]
    |> List.map (fun n ->
        let stopWatch = System.Diagnostics.Stopwatch.StartNew()
        let quad = logaritimicTime n
        stopWatch.Stop()
        (n, stopWatch.Elapsed.TotalMilliseconds))
    |> List.unzip

Chart.Line(x, y)
|> Chart.withXAxisStyle("numInput")
|> Chart.withYAxisStyle("milliseconds", MinMax = (0, 100))
|> Chart.withTitle("log n time")
|> Chart.withLineStyle(Width=2.0, Dash=StyleParam.DrawingStyle.Solid)
```
